# 数据收集

首先，利用Beautiful soup，爬取TED官网上的popular talks的第1~85页，获取所有播放量大于100万的演讲的演讲者、演讲时长、标题及播放页面的链接，再对进入每个演讲的播放页面，从播放页面上爬取该演讲的主题、浏览量和发布时间。

爬虫过程中，有时候会出现信息丢失，导致程序中止。当遇到找不到结点的情况时，先存入NaN，方便爬虫的继续和进一步的处理。

```python
class MySpider():
    name = "spider"

    def __init__(self, talks=None):

        self.headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.67 Safari/537.36'}
        #设置数据的存储
        self.title=[]
        self.speaker=[]
        self.time_duration=[]
        self.views=[]
        self.link=[]
        self.tag=[]
        self.date=[]

    def bs_request(self,url,parse):
        request = urllib.request.Request(url, headers=self.headers)
        html = urllib.request.urlopen(request).read()
        response = BeautifulSoup(html, 'html5lib')
        if parse==0:
            self.parse_homepage(url, response)
        else:
            self.parse_talk(url,response)

    # start函数，调用此函数可开始爬虫
    def start(self,urls,parse):
        for url in urls:
            self.bs_request(url,parse)
        if parse==0:
            print("homepage_over")
        else:
            print("talk_over")
            
    # 在ted的最受欢迎的演讲的主界面爬取受欢迎的演讲演讲者、标题、链接及演讲时长
    def parse_homepage(self,url,response):
    
        talks = response.find_all("div", class_="media__message")
        for talk in talks:
            # 提取每个演讲的演讲者
            speaker= talk.find("h4", class_="h12 talk-link__speaker").get_text()
            # 提取每个演讲的标题
            title =talk.find("a",class_="ga-link").get_text()
            title=title[1:-1]#删去头尾的\n
            self.speaker.append(speaker)
            self.title.append(title)
            #提取每个演讲的链接
            link=talk.find("a",class_="ga-link")
            self.link.append("https://www.ted.com"+link['href'])
      
        talks=response.find_all("div",class_="media__image media__image--thumb talk-link__image")
        for talk in talks:
            #提取每个演讲的时长 
            time_duration=talk.find("span",class_="thumb__duration").get_text()
            self.time_duration.append(time_duration)
    
    #在演讲的播放页面爬取演讲的主题、浏览量和发布时间
    def parse_talk(self,url,response): 
        
        #获取演讲的主题
        tags=response.find_all(attrs={"property":"og:video:tag"})
        tmp=[]
        for tag in tags:
            tmp.append(tag['content'])
        self.tag.append(tmp)
        
        #获取演讲的浏览量
        view=response.find(attrs={"itemprop":"interactionCount"})
        if view is not None:
            self.views.append(view['content'])
        else:
            print(url,"view_failed")
            self.views.append(np.nan)
                
        #获取演讲的发布时间    
        date=response.find(attrs={"itemprop":"uploadDate"})
        if date is not None:
            date=date['content']
            date=date[:10]
            self.date.append(date)
        else:
            print(url,"date_failed")
            self.date.append(np.nan)
            
        #某一演讲的相关数据爬取结束
        print("talk_over",url)
```

```python
#开始爬虫
spider=MySpider()
popular_urls=[]
talk_urls=[]
for i in range(1,85):
    popular_urls.append('https://www.ted.com/talks?page='+str(i)+'&sort=popular')
#从主界面爬取各个演讲的链接
spider.start(popular_urls,0)
talk_urls=spider.link
#从各个演讲的播放界面爬取相关信息
spider.start(talk_urls,1)
```

以下为数据转换过程，主要有四个转换工作：


①将网页显示的19:40和1h0m两种格式的演讲时长转换成以s为单位的时长。

```python
#定义演讲时长转换函数
def time_process(x):
    time=np.nan
    if x is not None:
        t=x.split(':')
        if len(t)==2:
            time= int(t[0])*60+int(t[1])
        else:
            hour,minute,quad=re.split(r"['h','m']", x)
            time=60*60*int(hour)+60*int(minute)
    return time

#进行转换
df['time duration']=df['time duration'].apply(lambda x:time_process(x))
```

②将浏览量从字符串类型转化为整数类型。

```python
df['view']=df['view'].astype(int)
```

③从演讲发布时间中提取出每个演讲的发布年份，生成新列。

```python
df['year']=df['date'].apply(lambda x:x.split('-')[0])
df['year']=df['year'].astype(int)
```

④从演讲发布时间中提取出每个演讲的发布月份，生成新列。

```python
df['month']=df['date'].apply(lambda x:x.split('-')[1])
df['month']=df['month'].astype(int)
```

# 基于机器学习的数据分析

```python
df_main=pd.read_csv('ted_main.csv')
df_main.head()
```

```python
import ast
df_main['tags']=df_main['tags'].apply(lambda x:ast.literal_eval(x))
s = df_main.apply(lambda x: pd.Series(x['tags']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'theme'
theme_df=df_main.drop('tags',axis=1).join(s)
theme_df.head()
```

```python
# coding: utf-8
import os
import jieba
from sklearn.model_selection import train_test_split     # 训练、测试集分类
from sklearn.preprocessing import LabelEncoder           # 标签方法
from tensorflow.keras.preprocessing.text  import Tokenizer          # 文本处理分词方法
from tensorflow.keras.preprocessing.sequence import pad_sequences   # 补零方法
from tensorflow.keras.utils import to_categorical          # 数值化分类方法
```

```python
text_df=theme_df[theme_df['theme'].isin(popular_theme.head(10)['theme'])][['theme','description']]
texts=text_df['description'].tolist()
categories=text_df['theme'].tolist()
n_sent=0
for i in range(len(texts)):
    n_sent=max(n_sent,len(texts[i]))
    seg=jieba.cut(texts[i])
    seg=' '.join(seg)
    texts[i]=seg
```

```python
# 总类别数
n_label = len(set(categories))
# One-Hot 方式对标签组重新打标签，形如：[[1.,0.,0.],[0.,1.,0.],[0.,0.,1.]]
encoder = LabelEncoder()
encoded_cate = encoder.fit_transform(categories)
label_cate = to_categorical(encoded_cate)

# 过滤标点符号以及特殊字符
tokenizer = Tokenizer(filters='。，、；：’“ ”—【】（）·★', split=' ')
tokenizer.fit_on_texts(texts)
# 对每个词进行word2id 编码
vocabs = tokenizer.word_index
# 文本编码后结果
seqs = tokenizer.texts_to_sequences(texts)
# One-Hot 方式对文本编码
oh_seqs = tokenizer.sequences_to_matrix(seqs)
# 对于没有达到最大特征长度的句子，左边补零
pad_oh_seqs = pad_sequences(oh_seqs, maxlen=n_sent)

# 对结果划分为训练集和测试集
X_trn, X_tst, y_trn, y_tst = train_test_split(pad_oh_seqs, label_cate, test_size=0.1, random_state=True)
```

```python
# 文本类数据通常使用Convolution1D，MaxPool1D
from tensorflow.keras.layers import Input, Dense, Dropout, Activation, Convolution1D, MaxPool1D, Concatenate, Flatten
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Model

# Embedding层神经元数量
EMB_SIZE = 300
```

```python
# 输入层
fin = Input(shape=(n_sent,), dtype='float32')

# 词向量Embedding层
embed = Embedding(len(vocabs) + 1, EMB_SIZE, input_length=n_sent)(fin)

# 1D卷积层-1
cnn1 = Convolution1D(filters=2, kernel_size=3, padding='same', strides=1, activation='relu')(embed)
# 1D Max Pooling层-1
mp1 = MaxPool1D(pool_size=4)(cnn1)

# 1D卷积层-2
cnn2 = Convolution1D(2, 4, padding='same', strides=1, activation='relu')(mp1)
# 1D Max Pooling层-2
mp2 = MaxPool1D(pool_size=4)(cnn2)

# 1D卷积层-3
cnn3 = Convolution1D(2, 5, padding='same', strides=1, activation='relu')(mp2)
# 1D Max Pooling层-3
mp3 = MaxPool1D(pool_size=4)(cnn3)

# 合并3个Max Pooling 层
cnn = Concatenate(axis=1)([mp1, mp2, mp3])
# 降维（展平）数据矩阵
flat = Flatten()(cnn)
# 丢弃20%神经元
dropout = Dropout(0.2)(flat)

# 全链接层
fout = Dense(n_label, activation='softmax')(dropout)
# 模型输出
model = Model(inputs=fin, outputs=fout)

# 编译模型
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 模型概要
model.summary()
```

```python
# 模型拟合，并在测试集上验证模型
model.fit(X_trn, y_trn, batch_size=128, epochs=30, validation_data=(X_tst, y_tst))
```

最终的结果并不理想，我认为最大的原因是演讲的简介并不能很好地突出演讲的主题，并且一个文本赋予了多个标签，对分类不能起到明确的监督效果，但若只用一个标签，需要人工筛选最合适的文本主题描述，而本身ted中演讲涵盖的内容实际上十分广泛，大多都涉及多个领域，没有特别明确的主题限定。故最终没有一个好的分类效果。
